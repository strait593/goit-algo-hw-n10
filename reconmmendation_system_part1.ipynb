```python
import pandas
import numpy
import matplotlib.pyplot as plt
from surprise import Dataset, accuracy
from surprise import SVD
from surprise.model_selection import cross_validate, train_test_split

#load the dataset
data = Dataset.load_builtin('ml-100k')

#train the model
train, test = train_test_split(data, train_size=0.8, test_size=0.2, random_state=42)
model_svd = SVD(n_factors=20, lr_all=0.005, reg_all=0.02)
model_svd.fit(train)
predictions = model_svd.test(test)
print(f"{accuracy.rmse(predictions=predictions):.2f}")

#cross validation
cross_validate(model_svd, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

### SVD++

```python
from surprise import SVDpp

#load the dataset
data = Dataset.load_builtin('ml-100k')
#train the model
train, test = train_test_split(data, train_size=0.8, test_size=0.2, random_state=42)
model_pp = SVDpp(n_factors=20, lr_all=0.005, reg_all=0.02)
model_pp.fit(train)
predictions = model_pp.test(test)
print(f"{accuracy.rmse(predictions=predictions):.2f}")
#cross validation
cross_validate(model_pp, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

### NMF

```python
from surprise import NMF

#load the dataset
data = Dataset.load_builtin('ml-100k')

#train the model
train, test = train_test_split(data, train_size=0.8, test_size=0.2, random_state=42)
model_nmf = NMF(n_factors=20)
model_nmf.fit(train)
predictions = model_nmf.test(test)
print(f"{accuracy.rmse(predictions=predictions):.2f}")

#cross validation
cross_validate(model_nmf, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
models = [model_svd, model_pp, model_nmf]

for result in models:
    print(f"accuracy: {result.__dict__}")
```

### Summary

NMF performed the best out of the bunch, and way faster than SVD++. I think the reason for that is that SVD++ is more complex and has more parameters to tune, which can lead to overfitting if not done properly. NMF, on the other hand, is a simpler model that can capture the underlying structure of the data without overfitting. Additionally, NMF can handle sparse data better than SVD++, which is often the case with recommendation systems.
